```python

class PhishingFeatureExtractor:
    def __init__(self, url=None):
        self.url = url

    # Function #2
    def url_length(self) -> int:
        return len(self.url) if self.url else 0


# --- Test cases ---
cases = [
    ("http://192.168.0.1/login", 24),
    ("https://example.com", 19),
    ("http://tiny.cc/x", 16),
    ("", 0),
    (None, 0),
]

# --- Run tests ---
for url, expected in cases:
    ex = PhishingFeatureExtractor(url)
    val = ex.url_length()
    status = "PASS" if val == expected else "FAIL"
    print(f"url={repr(url):30} -> length={val:3} (expected {expected:3}) [{status}]")

```
